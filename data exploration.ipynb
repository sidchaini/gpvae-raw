```python
i = 0
while i < 20:    
    sample = np.random.randint(0,x_train_full.shape[0])
    channel = np.random.randint(0,x_train_full.shape[2])

    if np.count_nonzero(np.isnan(x_train_full[sample,:,channel])) > 40:
        continue

    fig, ax1 = plt.subplots(1, 1, figsize=(5,5))
    
    ax1.plot(x_train_full[sample,:,channel])
    ax1.title.set_text('Full Data')

    
    plt.show()
#     plt.savefig(f"Fig_{i}.png",dpi=300)
    i+=1
```
